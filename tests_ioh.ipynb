```python
problem.reset()
bmax = -2147483648
bsol = None
sol = np.random.randint(0,2,800)
nbiter = 0
while nbiter < 10000 :
    value = problem(sol)
    if value >= bmax :
        bsol = sol
        bmax = value
    sol = np.random.randint(0,2,800)
    nbiter +=1
print(problem.state)
print(bsol)
```
